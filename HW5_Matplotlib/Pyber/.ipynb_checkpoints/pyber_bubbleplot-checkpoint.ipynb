```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)


# Combine the data into a single dataset
#merge the files pd.merge on=["city,"city"]

pyber_data_df = pd.merge(ride_data_df, city_data_df, on="city", how="left")

```

```python
# Display the data table for preview

pyber_data_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

urban_type = pyber_data_df[pyber_data_df["type"] == "Urban"].groupby([pyber_data_df["city"]])
suburban_type = pyber_data_df[pyber_data_df["type"] == "Suburban"].groupby([pyber_data_df["city"]])
rural_type = pyber_data_df[pyber_data_df["type"] == "Rural"].groupby([pyber_data_df["city"]])
urban_type.head()

#Urban Type Data Plots
rider_urban = urban_type["ride_id"].count()
fare_urban = urban_type["fare"].mean()
driver_urban = urban_type["driver_count"].mean()

#Suburban Type Data Plots

rider_suburban = suburban_type["ride_id"].count()
fare_suburban = suburban_type["fare"].mean()
driver_suburban = suburban_type["driver_count"].mean()

#Rural Type Data Plots

rider_rural = rural_type["ride_id"].count()
fare_rural = rural_type["fare"].mean()
driver_rural = rural_type["driver_count"].mean()


# Build the scatter plots for each city types
# matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, 
# norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, verts=None, edgecolors=None, *, data=None, **kwargs)[source]

plt.scatter(rider_urban, fare_urban, marker="o", facecolors="gold", edgecolors="black",
            s=driver_urban * 10, alpha=0.75, label = "Urban")
plt.scatter(rider_suburban, fare_suburban, marker="o", facecolors="lightskyblue", edgecolors="black",
            s=driver_suburban  * 10, alpha=0.75, label = "Suburban" )
plt.scatter(rider_rural, fare_rural, marker="o", facecolors="coral", edgecolors="black",
            s=driver_rural  * 10, alpha=0.75, label = "Rural" )


# Incorporate the other graph properties
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")

# Create a legend

legend = plt.legend(fontsize = 8, title= "City Types", loc="best")

# Incorporate a text label regarding circle size

plt.text(42,35,"Note: \nCircle size correlates with driver count per city.", fontsize = 10)

# Save Figure
plt.savefig("Pyber.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
type_groupped = pyber_data_df.groupby(['type'])
sumoffare = type_groupped['fare'].sum()

# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.1)
colors = ["coral", "lightskyblue", "gold"]
plt.pie(sumoffare, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")


# Save Figure
```

```python
# Show Figure
plt.show()
```
